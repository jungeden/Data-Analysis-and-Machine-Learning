### 6.3

```python
df_1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index = ['0', '1', '2'])

df_2 = pd.DataFrame({
    'B': ['B2', 'B3'],
    'C': ['C2', 'C3'],
    'D': ['D2', 'D3'],
    'E': ['E2', 'E3']
}, index = ['0', '1'])

print(f"{df_1}\n\n{df_2}\n")
```

#### 1)

```python
df_3 = pd.concat([df_1, df_2])
print(df_3)
```

#### 2)

```python
df_4 = pd.concat([df_1, df_2], axis=1, join='inner')
print(df_4)
```

### 6.4

```python
path = 'http://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'trade_month.csv'
trade = pd.read_csv(file, encoding='cp949')
```

#### 1)

```python
trade.head(5)
```

#### 2)

```python
trade.info()
```

#### 3)

```python
trade['기간'] = trade['기간'].astype(str)

for x in list(trade.columns[2:]):
    trade[x] = trade[x].str.replace(',','').astype(float)

trade.info()
```

#### 4)

```python
print(trade.isnull().sum)
```

#### 5)

```python
trade[trade.isnull().any(axis=1)]
```

#### 6)

```python
trade567 = trade[(trade['국가명']=='체코공화국') & (trade['기간']>='2015.05') & (trade['기간']<='2015.07') & (trade['기간']!='2015.06')]

trade.loc[1189, '수출금액'] = trade567['수출금액'].mean()
trade.loc[1189, '무역수지'] = trade567['무역수지'].mean()

trade.loc[[1189]]
```

#### 7)

```python
display(trade[trade.duplicated()])
display(trade[(trade['기간']=='2015.02')&(trade['국가명']=='중국')])
trade.drop_duplicates(inplace=True)
display(trade[trade.duplicated()])
```
