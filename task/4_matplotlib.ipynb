```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 10)
y = np.linspace(0, np.pi, 10)

X, Y = np.meshgrid(x, y)
data = np.sin(X) + np.sin(Y)

plt.imshow(data)
plt.colorbar()
plt.title('1) result')
plt.show()
```

### 4.3

```python
import numpy as np
import matplotlib.pyplot as plt

grid = plt.GridSpec(2,2,wspace=0.2,hspace=0.2)

f1 = np.random.normal(loc=0, scale=1, size=10000)
f2 = np.random.normal(loc=3, scale=.5, size=10000)

plt.subplot(grid[0,0])
plt.title('Red histogram')
plt.hist(f1, bins=200, color='r',alpha=.4)
plt.margins(0.1)
plt.legend(frameon=False)

plt.subplot(grid[0,1])
plt.title('Blue histogram')
plt.hist(f2, bins=200, color='b',alpha=.4)
plt.margins(0.1)
plt.legend(frameon=False)


plt.subplot(grid[1,0:])
plt.hist(f1, bins=200, alpha=.4, label='loc=0, scale=1', color='r' )
plt.hist(f2, bins=200, alpha=.4, label='loc=3, scale=.5',color='b')
plt.margins(0.1)
plt.legend(loc='upper left')

```

### 4.4

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) 
plt.figure(figsize=(15,5))
grid = plt.GridSpec(1,3,wspace=0.2,hspace=0.2)

x = np.random.normal(25, 6, 1000)
y = np.random.normal(25, 6, 1000)

plt.subplot(grid[0,0])
plt.title('x: N(25,6), y: N(25,6)')
plt.scatter(x, y, color='b')
plt.xlabel('x')
plt.ylabel('y') 
plt.xlim(0,50)
plt.ylim(0,50)


x = np.random.normal(25, 6, 1000)
y = np.random.normal(25, 3, 1000)

plt.subplot(grid[0,1])
plt.title('x: N(25,6), y: N(25,3)')
plt.scatter(x, y, color='r')
plt.xlabel('x')
plt.ylabel('y') 
plt.xlim(0,50)
plt.ylim(0,50)


x = np.random.normal(25, 6, 1000)
y = x + np.random.normal(0, 1, 1000)

plt.subplot(grid[0,2])
plt.title('x: N(25,6), y: x + N(0,1)')
plt.scatter(x, y, color='g')
plt.xlabel('x')
plt.ylabel('y') 
plt.xlim(0,50)
plt.ylim(0,50)

```

### 4.5

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(30):
    x = np.random.randint(0,100)
    y = np.random.randint(0,100)

    plt.scatter(x,y,color='b')

   

plt.title('Scatter Plot of Random Coordinates')
plt.grid(True, alpha=.7)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xlabel('x-coordinate')
plt.ylabel('y-coordinate')
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
grid = plt.GridSpec(1,3,wspace=0.2,hspace=0.2)

d = [
    [[77,25], [78,28], [85,19], [83,30], [73,21], [77,22], [73,17], [80,35]], 
    [[75,56], [77,57], [86,50], [86,53], [79,60], [83,53], [83,49], [88,61]], 
    [[34,22], [38,25], [38,19], [41,30], [30,21], [37,24], [41,28], [35,18]]
    ]

dtitle = ['Dachshund size', 'Samoyed size', 'Maltese size']
dcolor = ['r', 'b', 'g']
dmarker = ['o', 's', '^']

for i in range(3):
    data = np.array(d[i])
    plt.subplot(grid[0,i])
    plt.title(dtitle[i])
    
    plt.scatter(data[:,0], data[:,1], color=dcolor[i], marker=dmarker[i])
    plt.xlabel('Length')
    plt.ylabel('Height')
    plt.margins(0.1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

d = [
    [[77,25], [78,28], [85,19], [83,30], [73,21], [77,22], [73,17], [80,35]], 
    [[75,56], [77,57], [86,50], [86,53], [79,60], [83,53], [83,49], [88,61]], 
    [[34,22], [38,25], [38,19], [41,30], [30,21], [37,24], [41,28], [35,18]]
    ]

dlabel = ['Dachshund', 'Samoyed', 'Maltese']
dcolor = ['r', 'b', 'g']
dmarker = ['o', 's', '^']


for i in range(3):
    data = np.array(d[i])
    plt.scatter(data[:,0], data[:,1], color=dcolor[i], marker=dmarker[i], label=dlabel[i])

plt.title('Dog size')
plt.xlabel('Length')
plt.ylabel('Height')
plt.margins(0.1)
plt.legend()
```
