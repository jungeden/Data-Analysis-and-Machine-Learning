```python
line = "*"*40
first = "Hello! I will introduce myself"
info = [["Jeong sejin", "Changwon", "23", "Iksan", "Student"], ["Kim seobin", "jeonju", "23", "Jeonju", "Student"], ["cha danwoo", "Seoul", "23", "Jeonju", "Student"]]
print(line)
for person in info:
    print(first)
    for data in person:
        print(data)
    print(line)
```
