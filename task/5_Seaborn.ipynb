## Chapter 5. 통계 데이터와 시본 라이브러리

### 5.1

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.set_theme(style="whitegrid")

sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"
)
plt.show()
```

### 5.2

#### 1)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.countplot(
    data = titanic,
    x="class",
    hue="who"
)
plt.show()
```

#### 2)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.relplot(
    data = titanic,
    x="age",
    y="fare",
    col="survived"
)
plt.show()
```
