```python
import pandas as pd
import numpy as np

month_se = pd.Series(['1월', '2월', '3월', '4월'])
income_se = pd.Series([9500, 6200, 6050, 7000])
expenses_se = pd.Series([5040, 2350, 2300, 4800])
df = pd.DataFrame( {'월': month_se, '수익': income_se,
                   '지출' : expenses_se})

print(f'동윤이네 상점의 수익 합계:{income_se.sum():,}')
print(f'동윤이네 상점의 지출 합계:{expenses_se.sum():,}')
```

```python
import pandas as pd
import numpy as np

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path+'vehicle_prod.csv'
df = pd.read_csv(file, index_col = 0)

sum_row = df.select_dtypes(np.number).sum().rename('total')
todf = pd.concat([df, sum_row.to_frame().T])

todf
```
