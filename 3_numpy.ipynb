## Chapter 3. 파이썬 최고의 라이브러리: 넘파이

### 3.1

#### 1)

```python
import numpy as np

a = np.arange(1,46)
a
```

#### 2)

```python
np.random.seed(85)
np.random.shuffle(a)

lotto = a[0:6]
print(f"lotto: {lotto}")
```

#### 3)

```python
lotto.sort()
new = a[7]

print(f"lotto: {lotto}, 추가번호: {new}")
```

#### 4)

```python
n_arr = np.arange(25).reshape(5,5)
print(n_arr)
```

#### 5)

```python
print(f"첫 원소: {n_arr[0][0]}\n마지막 원소: {n_arr[-1][-1]}\n")
```

#### 6)

```python
n_arr2 = n_arr[0:2, 0:]
print(n_arr2)
```

#### 7)

```python
n_arr3 = n_arr[0: , ::2]
print(n_arr3)
```

#### 8)

```python
n_arr4 = n_arr[0:2,0:].reshape(-1)
n_arr4 = np.reshape(n_arr4, (5, 2))
print(f"{n_arr4}")
```
